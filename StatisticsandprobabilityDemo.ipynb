### 2.Poisson Distribution

```python
from scipy.stats import poisson
```

```python
poisson.pmf(2,3)  #Probability mass function at k of the given RV,X=2,Mean=3
```

2.1 Suppose bank customers arrive randomly on weekday afternoons at an average of 3.2 customers every 4 minutes. What is the probability of exactly 5 customers arriving in a 4-minute interval on a weekday afternoon? 

```python
poisson.pmf(5,3.2)  #Probability mass function at k of the given RV,X=5,mean=3.2
```

Q2.2 Bank customers arrive randomly on weekday afternoons at an average of 3.2 customers every 4 minutes. What is the probability of having more than 7 customers in a 4-minute interval on a weekday afternoon? 

```python
prob=poisson.cdf(7,3.2)  #Cumulative distribution function of the given RV,X=7,mean=3.2
prob
```

```python
prob_more_than_7=1-prob
```

prob_more_than_7

Q2.3, A bank has an average random arrival rate of 3.2 customers every 4 minutes. What is the probability of getting exactly 10 customers during an 8-minute interval? 

```python
poisson.pmf(10,6.4)
```
